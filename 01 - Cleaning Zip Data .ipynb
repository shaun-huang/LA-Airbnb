## Import Data and Modules

```python
import pandas as pd
```

## read zip file

```python
# Read zip file
df = pd.read_csv("Data/uszips.csv")
df.head()
```

## Data Cleaning

### a) Drop Useless Columns

```python
drop_columns_df = df.drop(columns=[
    "zcta", #not needed
    "parent_zcta", #not needed
    "county_fips", #not needed
    "all_county_weights", #not needed
    "imprecise", #not needed
    "military", #not needed
    "timezone", #not needed
])
drop_columns_df.head()
```

### a) Drop Useless Columns

```python
rename_columns_df = drop_columns_df.rename(columns={
    "lat":"zip_lat",
    "lng":"zip_lng"
})

rename_columns_df.head()
```

## Export Clean Zip file

```python
rename_columns_df.to_csv("Data/clean_zip.csv",index=False)
```
