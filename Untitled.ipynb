```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "Data/clean_data.csv"

data = pd.read_csv(path)
data.head()
```

```python
data.drop(data.columns[[0]], axis=1, inplace=True)
data.head()
```

```python
# GroupBy Ratings and City 
ratings = data[["id","city_y","review_scores_rating","property_type"]]
ratings.head()
```

```python
city_rating=ratings.dropna()
```

```python
city_rating.head()
```

```python
#Taking the average of scores in Los Angeles 
df1=city_rating.loc[:,["id","city_y","review_scores_rating"]]
la_score = df1[df1.city_y == "Los Angeles"]
la_score.head()
```

```python
score_df=city_rating.groupby(['city_y']).median().sort_values('review_scores_rating',ascending=True)[["review_scores_rating"]]
score_df
```

```python
score_df.to_csv('Data/score_df')
```
