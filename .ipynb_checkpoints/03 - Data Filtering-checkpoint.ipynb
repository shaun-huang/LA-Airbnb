```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "Data/clean_data.csv"

data = pd.read_csv(path)
data.head()
```

```python
data.drop(data.columns[[0]], axis=1, inplace=True)
data.head()
```

```python
# GroupBy City
city = data[["id", "city_y", "zip"]]
city.head()
```

```python
# Count Each City
count = city["city_y"].value_counts()
count
```

```python
count_df = pd.DataFrame(count)
count_df
```

```python
# Total Count

total_cities_name = len(city["city_y"].value_counts())
total_cities_count = city["city_y"].count()



# Create new DataFrame
city_analysis = pd.DataFrame({"Number of City": [total_cities_name],
                               "Total Listing": [total_cities_count]})
city_analysis
```
