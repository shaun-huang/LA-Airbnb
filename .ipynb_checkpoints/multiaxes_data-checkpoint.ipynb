```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "Data/clean_data.csv"

axes_data = pd.read_csv(path)
axes_data.head()
```

```python
axes_data.drop(axes_data.columns[[0]], axis=1, inplace=True)
axes_data.head()
```

```python
axes_df=axes_data.dropna()
axes_df.head()
```

```python
axes_df1=axes_df.groupby(['city_y']).median().sort_values('log_price',ascending=True)
axes_df1
```

```python
axes_df1.to_csv('Data/axes_df.csv')
```

```python
axes_df2=axes_df.groupby(['property_type']).median().sort_values('number_of_reviews',ascending=False)
axes_df2
```

```python
axes_df2.to_csv('Data/multiaxes_df.csv')
```
