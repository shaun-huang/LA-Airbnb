```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "Data/clean_data.csv"

axes_data = pd.read_csv(path)
axes_data.head()
```

```python
axes_data.drop(axes_data.columns[[0]], axis=1, inplace=True)
axes_data.head()
```

```python
axes_df=axes_data.dropna()
axes_df.head()
```

```python
df1 = axes_df[["city_y","review_scores_rating","property_type","neighbourhood","number_of_reviews","host_since","room_type","log_price","bedrooms","bathrooms"]]
df1
```

```python
df2=df1.sort_values('host_since',ascending=True)
df2.dropna().head()
```

```python
#df2['year']=pd.DatetimeIndex(df2['host_since']).year
#df2.head()
```

```python
la_df=df2.loc[df2['city_y']=="Los Angeles"]
la_df
```

```python
la_df2=la_df[["neighbourhood","property_type","host_since","log_price"]]
la_df2
```

```python
la_df3=la_df2.groupby(['neighbourhood']).mean().sort_values('log_price',ascending=False)
la_df3
```

```python
la_df2.to_csv("Data/bedbath.csv")
```

```python
la_df3=la_df2.drop(["Yurt"])
la_df3.to_csv("Data/pricetype.csv")
```
