```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "Data/clean_data.csv"

axes_data = pd.read_csv(path)
axes_data.head()
```

```python
axes_data.drop(axes_data.columns[[0]], axis=1, inplace=True)
axes_data.head()
```

```python
axes_df=axes_data.dropna()
axes_df.head()
```

```python
axes_df[axes_df['neighbourhood'] == "Rolling Hills Estates"]
```

```python
axes_df1=axes_df.groupby(['neighbourhood']).mean().sort_values('number_of_reviews',ascending=False)
axes_df1
```

```python
axes_df1.to_csv('Data/axes_df.csv')
```

```python
axes_df2=axes_df.groupby(['property_type']).mean().sort_values('number_of_reviews',ascending=False)
axes_df2
```

```python
multiaxes_df=axes_df2.drop(["Chalet"])
multiaxes_df
```

```python
multiaxes_df.to_csv('Data/multiaxes_df.csv')
```
