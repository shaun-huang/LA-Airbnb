```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "Data/clean_data.csv"

data = pd.read_csv(path)
data.head()
```

```python
data.drop(data.columns[[0]], axis=1, inplace=True)
data.head()
```

```python
# GroupBy Ratings and City and and Property type 
ratings = data[["id","city_y","review_scores_rating","property_type","population","neighbourhood","number_of_reviews"]]
ratings.head()
```

```python
city_rating=ratings.dropna()
```

```python
city_rating.head()
```

```python
#Taking the average of scores in Los Angeles 
#df1=city_rating.loc[:,["id","city_y","review_scores_rating"]]
#la_score = df1[df1.city_y == "Los Angeles"]
#la_score.head()
```

```python
#Median score ratings across cities 
score_df=city_rating.groupby(['neighbourhood']).mean().sort_values('review_scores_rating',ascending=False)
score_df
```
